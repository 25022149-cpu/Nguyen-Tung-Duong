```python
#W4A6
import sys

MAX = 1000005
is_prime = [True] * MAX
is_prime[0] = is_prime[1] = False

# Sàng nguyên tố Eratosthenes
for p in range(2, int(MAX**0.5) + 1):
    if is_prime[p]:
        for i in range(p * p, MAX, p):
            is_prime[i] = False

# Mảng cộng dồn
pref_sum = [0] * MAX
for i in range(1, MAX):
    pref_sum[i] = pref_sum[i - 1] + (i if is_prime[i] else 0)

# Đọc toàn bộ dữ liệu một lần (tránh lỗi ngắt dòng không đều)
input_data = sys.stdin.read().split()

# Duyệt từng cặp a, b
for i in range(0, len(input_data) - 1, 2):
    a = int(input_data[i])
    b = int(input_data[i+1])
    print(pref_sum[b] - pref_sum[a - 1])
    
```

```python
#4A8
n = int(input())
steps = 0

# Ép sang chuỗi (str) để kiểm tra đối xứng dễ dàng hơn
while str(n) != str(n)[::-1]:
    # Chuỗi [::-1] dùng để đảo ngược chuỗi trong Python
    n += int(str(n)[::-1])
    steps += 1

print(f"{steps} {n}")
```

```python
#4A9
def has_distinct_digits(num):
    s = str(num)
    # set() sẽ loại bỏ các phần tử trùng lặp
    return len(set(s)) == len(s)

try:
    n = int(input())
    results = []
    i = 1
    while i * i <= n:
        sq = i * i
        if has_distinct_digits(sq):
            results.append(str(sq))
        i += 1
    
    if results:
        print(" ".join(results))
    else:
        print("no number")
except EOFError:
    pass
```

```python
#4A10
def get_collatz_len(num):
    length = 1
    while num > 1:
        if num % 2 == 0:
            num //= 2
        else:
            num = 3 * num + 1
        length += 1
    return length

n = int(input())
max_len = 0
best_x = 0

# Duyệt từ 1 đến n, vòng lặp này đảm bảo nếu cùng độ dài, số nhỏ hơn sẽ được giữ lại
for x in range(1, n + 1):
    curr_len = get_collatz_len(x)
    if curr_len > max_len:
        max_len = curr_len
        best_x = x

print(f"{best_x} {max_len}")
```

```python
#4A11
import math

n = int(input())

if n % 2 != 0:
    print(0)
else:
    count = 0
    for i in range(1, int(math.isqrt(n)) + 1):
        if n % i == 0:
            # Kiểm tra ước i
            if i % 2 == 0:
                count += 1
            # Kiểm tra ước còn lại n/i
            other = n // i
            if other != i and other % 2 == 0:
                count += 1
    print(count)
```

```python
#4A12
def calculate_interest():
    try:
        x = float(input("Nhập số tiền gửi ban đầu (X): "))
        n = int(input("Nhập số tháng gửi (N): "))
        
        # Lãi suất 0.7% mỗi tháng
        rate = 0.007 
        
        # Công thức tính lãi kép
        total_amount = x * ((1 + rate) ** n)
        
        # In ra số tiền, bỏ qua phần lẻ thập phân
        print(int(total_amount))
        
    except ValueError:
        print("Vui lòng nhập một số hợp lệ.")

# Gọi hàm để chạy thử
if __name__ == "__main__":
    calculate_interest()
```

```python
#4A13
def sum_of_proper_divisors(num):
    if num <= 1:
        return 0
    total = 1 # 1 luôn là ước của mọi số nguyên dương
    
    # Chỉ cần chạy vòng lặp đến căn bậc 2 của num để tối ưu thời gian
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            total += i
            # Tránh cộng trùng nếu i là căn bậc hai của num (ví dụ: num = 25, i = 5)
            if i != num // i:
                total += num // i
    return total

def check_amicable_numbers():
    try:
        a = int(input("Nhập số thứ nhất: "))
        b = int(input("Nhập số thứ hai: "))
        
        # Kiểm tra điều kiện số thân thiết
        if sum_of_proper_divisors(a) == b and sum_of_proper_divisors(b) == a:
            print("true")
        else:
            print("false")
            
    except ValueError:
        print("Vui lòng nhập số nguyên hợp lệ.")

# Gọi hàm để chạy thử
if __name__ == "__main__":
    check_amicable_numbers()
```

```python
#4A14
def find_gcd():
    try:
        m = int(input("Nhập số nguyên dương m: "))
        n = int(input("Nhập số nguyên dương n: "))
        
        # Thuật toán Euclid tìm Ước chung lớn nhất
        while n != 0:
            m, n = n, m % n
            
        print(m)
        
    except ValueError:
        print("Vui lòng nhập số nguyên hợp lệ.")

# Gọi hàm để chạy thử
if __name__ == "__main__":
    find_gcd()
```

```python
#4A15
def solve_chicken_and_dog():
    try:
        total_animals = int(input("Nhập tổng số con: "))
        total_legs = int(input("Nhập tổng số chân: "))
        
        # Kiểm tra tính hợp lệ sơ bộ: 
        # - Số chân phải là số chẵn
        # - Số chân tối thiểu (nếu toàn gà) là 2 * tổng số con
        # - Số chân tối đa (nếu toàn chó) là 4 * tổng số con
        if total_legs % 2 != 0 or total_legs < 2 * total_animals or total_legs > 4 * total_animals:
            print("invalid")
            return
            
        # Tính toán số lượng chó và gà dựa trên hệ phương trình
        dogs = (total_legs - 2 * total_animals) // 2
        chickens = total_animals - dogs
        
        # Đảm bảo số lượng không bị âm
        if chickens >= 0 and dogs >= 0:
            print(f"Gà: {chickens}")
            print(f"Chó: {dogs}")
        else:
            print("invalid")
            
    except ValueError:
        # Xử lý trường hợp người dùng nhập chữ thay vì số
        print("invalid")

# Gọi hàm để chạy thử
if __name__ == "__main__":
    solve_chicken_and_dog()
```
