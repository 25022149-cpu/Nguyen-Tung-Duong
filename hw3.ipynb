```python
#4
h = int(input("Bài 4 - Nhập vào độ cao h: "))

for i in range(1, h + 1):
    # In khoảng trắng lùi vào đầu dòng
    print(" " * (h - i), end="")
    
    if i == 1:
        # Dòng đầu tiên chỉ có 1 ngôi sao
        print("*")
    elif i == h:
        # Dòng cuối cùng in đầy sao
        print("*" * (2 * i - 1))
    else:
        # Các dòng ở giữa in rỗng
        print("*", end="")
        print(" " * (2 * i - 3), end="")
        print("*")
```

```python
#9
a = int(input("Nhập a: "))
is_chinh_phuong = False
for i in range(1, a + 1):
    if i * i == a:
        is_chinh_phuong = True
        break
    if i * i > a: # Tối ưu: nếu bình phương vượt quá a thì dừng
        break

if is_chinh_phuong:
    print(f"{a} là số chính phương")
else:
    print(f"{a} không phải số chính phương")
```

```python
#12
n = int(input("Nhập n: "))
s_k = 0
k = 0

while s_k + (k + 1) < n:
    k += 1
    s_k += k

print(f"Số k tìm được là: {k}")
print(f"Khi đó S({k}) = {s_k} (nhỏ hơn {n})")
```

```python
#13
def find_min_n():
    try:
        a = float(input("Nhập số nguyên dương A: "))
        total_sum = 0
        n = 0
        
        # Vòng lặp chạy cho đến khi tổng vượt quá A
        while total_sum <= a:
            n += 1
            total_sum += 1 / n
            
        print(f"Số n nhỏ nhất tìm được là: {n}")
        print(f"Tổng thực tế: {total_sum}")
        
    except ValueError:
        print("Vui lòng nhập một số hợp lệ.")

if __name__ == "__main__":
    find_min_n()
```

```python
#14
def find_max_min_no_list():
    print("Nhập dãy số nguyên (nhập -1 để dừng):")
    
    max_val = None
    min_val = None
    
    while True:
        try:
            num = int(input("Nhập số: "))
            
            # Điều kiện dừng
            if num == -1:
                break
            
            # Nếu là số đầu tiên, gán nó cho cả max và min
            if max_val is None:
                max_val = num
                min_val = num
            else:
                # Cập nhật Max
                if num > max_val:
                    max_val = num
                # Cập nhật Min
                if num < min_val:
                    min_val = num
                    
        except ValueError:
            print("Vui lòng nhập số nguyên.")

    # Kiểm tra xem người dùng có nhập số nào trước khi nhấn -1 không
    if max_val is not None:
        print(f"Số lớn nhất: {max_val}")
        print(f"Số nhỏ nhất: {min_val}")
    else:
        print("Dãy số trống.")

if __name__ == "__main__":
    find_max_min_no_list()
```

```python
#15
n = int(input("Nhập n: "))
temp = abs(n) # Dùng trị tuyệt đối để xử lý cả số âm
count = 0
if temp == 0: count = 1
while temp > 0:
    count += 1
    temp //= 10
print(f"Số chữ số: {count}")
```

```python
#16
n = int(input("Nhập n: "))
chan = 0
le = 0
temp = abs(n)
if temp == 0: chan = 1

while temp > 0:
    chu_so = temp % 10
    if chu_so % 2 == 0:
        chan += 1
    else:
        le += 1
    temp //= 10

print(f"Chữ số chẵn: {chan}, Chữ số lẻ: {le}")
```

```python
#18
n = int(input("Nhập n: "))
is_power_of_3 = True

if n <= 0:
    is_power_of_3 = False
else:
    temp = n
    while temp > 1:
        if temp % 3 != 0:
            is_power_of_3 = False
            break
        temp //= 3

print(f"{n} có phải dạng 3^k: {is_power_of_3}")
```

```python
#19
a = int(input("Nhập A: "))
f1, f2 = 1, 1
fib_max = 1

while f2 <= a:
    fib_max = f2
    f1, f2 = f2, f1 + f2

print(f"Số Fibonacci lớn nhất không vượt quá {a} là: {fib_max}")
```

```python
#20
n = int(input("Nhập n: "))
count = 0
temp = abs(n)
while temp > 0:
    chu_so = temp % 10
    if chu_so in [2, 3, 5, 7]:
        count += 1
    temp //= 10
print(f"Số lượng chữ số là số nguyên tố: {count}")
```
