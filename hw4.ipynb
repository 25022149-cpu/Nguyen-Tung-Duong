```python
#23
n = int(input("Nhập n: "))

S = 0
k = 0
while S + (k + 1) < n:
    k += 1     
    S += k      

print(f"Giá trị k tìm được là: {k}")
print(f"Khi đó tổng S({k}) = {S}, nhỏ hơn {n}")
```

```python
#24
def find_min_n():
    try:
        a = float(input())
        sum = 0
        n = 0
        
        while sum <= a:
            n += 1
            sum += 1 / n
            
        print(f"Số n nhỏ nhất tìm được là: {n}")
        print(f"Tổng thực tế: {sum}")
        
    except ValueError:
        print("Vui lòng nhập một số hợp lệ.")

if __name__ == "__main__":
    find_min_n()
```

```python
#25

a = input().split()

max_val = None
min_val = None

for x in a:
    n = int(x)
    if n == -1:
        break
    
    if max_val is None:
        max_val = min_val = n
    else:
        if n > max_val: max_val = n
        if n < min_val: min_val = n

if max_val is not None:
    print(f"Lớn nhất: {max_val}, Nhỏ nhất: {min_val}")
else:
    print("Không có dữ liệu.")
```

```python
#26
a = int(input("Nhập A: "))
f1, f2 = 1, 1
fib_max = 1

while f2 <= a:
    fib_max = f2
    f1, f2 = f2, f1 + f2

print(f"Số Fibonacci lớn nhất không vượt quá {a} là: {fib_max}")
```

```python
#29
n = input("Nhập chuỗi: ")
a = ""

for ky_tu in n:
    
    if ky_tu.islower():
        a += ky_tu.upper()
   
    elif ky_tu.isupper():
        a += ky_tu.lower()
    
    else:
        a += ky_tu

print("Chuỗi sau khi đổi:", a)
```

```python
#31
s = input()

count_hoa = 0
count_thuong = 0
count_so = 0

for i in range(len(s)):
    ky_tu = s[i]
    if ky_tu.isupper():
        count_hoa += 1
    elif ky_tu.islower():
        count_thuong += 1
    elif ky_tu.isdigit():
        count_so += 1

print("Số ký tự in hoa:", count_hoa)
print("Số ký tự in thường:", count_thuong)
print("Số ký tự số:", count_so)
```
